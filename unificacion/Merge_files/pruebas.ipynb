# Data Cleaning

```python
import os

path_carpeta = 'data/desordenadas/2023-11-29.v1i.yolov7pytorch/'

mapeo_carpeta = {
    '0':'0',
    '1':'22',
    '2':'9',
    '3':'11',
    '4':'5',
    '5':'17'
}
```

```python
# funcion para actualizar las etiquetas de clase
def actualizar_etiquetas(lineas, mapeo_clases):
    etiquetas_actualizadas = []
    for linea in lineas:
        partes = linea.strip().split()
        if partes:
            clase_original = partes[0]
            if clase_original in mapeo_clases:
                partes[0] = mapeo_clases[clase_original]
            etiqueta_actualizada = ' '.join(partes)
            etiquetas_actualizadas.append(etiqueta_actualizada + '\n')
    return etiquetas_actualizadas
```

```python
# funcion para abrir - actualizar - cerrar un archivo
vacias = {}
for folder in os.listdir(path_carpeta):
    folder_path = os.path.join(path_carpeta, folder)
    if os.path.isdir(folder_path):
        print(f"Carpeta: {folder}")
        images_path = os.path.join(folder_path, 'images')
        labels_path = os.path.join(folder_path, 'labels')
        for filename in os.listdir(labels_path):

            with open(os.path.join(labels_path, filename), 'r') as f:
                lines = f.readlines()
            
            if not lines:
                if folder not in vacias:
                    vacias[folder] = []
                vacias[folder].append(filename)
                continue

            etiquetas_actualizadas = actualizar_etiquetas(lines, mapeo_carpeta)

            with open(os.path.join(labels_path, filename), 'w') as f:
                f.writelines(etiquetas_actualizadas)
```

```python
# delete empty files
cont = 0
for folder in vacias:
    for filename in vacias[folder]:
        print(f"Eliminando {filename.replace('.txt', '')} de {folder} en labels e images en {path_carpeta}")
        os.remove(os.path.join(path_carpeta, folder, 'labels', filename))
        os.remove(os.path.join(path_carpeta, folder, 'images', filename.replace('.txt', '.jpg')))
        cont += 1
```

```python
print(f"Se han eliminado {cont} archivos vacíos")
```

# Funcion que unifica todos los archivos

```python
import os
import shutil

def merge_folders(root_path, root_destino, subfolder_names):
    # Crear las carpetas de destino si no existen
    destination_folders = {name: os.path.join(root_destino, name) for name in subfolder_names}
    for dest_folder in destination_folders.values():
        os.makedirs(dest_folder, exist_ok=True)
        # Crear subcarpetas 'images' y 'labels'
        os.makedirs(os.path.join(dest_folder, 'images'), exist_ok=True)
        os.makedirs(os.path.join(dest_folder, 'labels'), exist_ok=True)

    # Recorrer cada carpeta de fecha
    date_folders = [f for f in os.listdir(root_path) if os.path.isdir(os.path.join(root_path, f))]

    for date_folder in date_folders:
        for subfolder_name in subfolder_names:
            # Ruta de la subcarpeta actual
            current_subfolder_path = os.path.join(root_path, date_folder, subfolder_name)

            # Dividir en imágenes y etiquetas
            for subsubfolder_name in ['images', 'labels']:
                subsubfolder_path = os.path.join(current_subfolder_path, subsubfolder_name)
                if os.path.exists(subsubfolder_path):
                    # Copiar todos los archivos de esta subcarpeta a la carpeta de destino correspondiente
                    for filename in os.listdir(subsubfolder_path):
                        src_file = os.path.join(subsubfolder_path, filename)
                        dst_file = os.path.join(destination_folders[subfolder_name], subsubfolder_name, filename)
                        shutil.copy2(src_file, dst_file)

root_dir = 'data/desordenadas/'  # Cambia esto por la ruta de tu directorio
root_dir_destino = 'data/juntas/'
subfolders = ['train', 'test', 'valid']
merge_folders(root_dir, root_dir_destino, subfolders)
```

# Revisar carpeta YOLO

```python
import os

def revisar_carpeta_yolo(path_base):
    for tipo in ['train', 'val', 'test']:
        carpeta_images = os.path.join(path_base, tipo, 'images')
        carpeta_labels = os.path.join(path_base, tipo, 'labels')
        verificar_duplicados = [f for f in os.listdir(carpeta_labels) if f.endswith('.txt')]

        archivos_images = {f.split('.')[0] for f in os.listdir(carpeta_images) if f.endswith(('.jpg', '.png'))}
        archivos_labels = {f.split('.')[0] for f in os.listdir(carpeta_labels) if f.endswith('.txt')}

        # Verificar si cada imagen tiene su correspondiente archivo de etiqueta
        for archivo in archivos_images:
            if archivo not in archivos_labels:
                print(f"Falta archivo de etiqueta para la imagen {archivo} en {tipo}")
        
        for archivo in archivos_labels:
            if archivo not in archivos_images:
                print(f"Falta imagen para el archivo de etiqueta {archivo} en {tipo}")

        # Verificar duplicados
        for archivo in verificar_duplicados:
            ruta_archivo = os.path.join(carpeta_labels, archivo)
            with open(ruta_archivo, 'r') as f:
                lineas = f.read().splitlines()

            if len(lineas) != len(set(lineas)):
                print(f'Etiquetas duplicadas encontradas en {ruta_archivo}')



path_base = 'data/juntas'
revisar_carpeta_yolo(path_base)
```

```python
import os
from collections import defaultdict

def contar_etiquetas_por_clase(path_base, nombres_clases):
    contador_clases = defaultdict(int)
    
    for tipo in ['train', 'val', 'test']:
        carpeta_labels = os.path.join(path_base, tipo, 'labels')
        
        # Contar etiquetas por clase
        for archivo_label in os.listdir(carpeta_labels):
            if archivo_label.endswith('.txt'):
                ruta_archivo_label = os.path.join(carpeta_labels, archivo_label)
                with open(ruta_archivo_label, 'r') as f:
                    for linea in f:
                        if linea.strip():  # Asegurarse de que la línea no esté vacía
                            clase = int(linea.split()[0])  # Convertir a entero para obtener el índice de la clase
                            nombre_clase = nombres_clases[clase]  # Obtener el nombre de la clase
                            contador_clases[nombre_clase] += 1

    # Ordenar clases por conteo de menor a mayor
    clases_ordenadas = sorted(contador_clases.items(), key=lambda item: item[1])

    # Imprimir conteo por clase ordenado
    for nombre_clase, conteo in clases_ordenadas:
        print(f'Clase {nombre_clase}: {conteo} etiquetas')

# Nombres de las clases según su índice
nombres_clases = ['Bandeja', 'Carne de cerdo', 'Chorizo', 'Fideos', 'Garbanzos', 'Plato', 'Arroz', 'Hamburguesa', 
                  'Carne molida', 'Budin de verduras', 'Ensaladas', 'Pescado frito', 'Pollo arvejado', 'Pollo asado', 
                  'Huevo', 'Charquican', 'Lentejas', 'Pure', 'Vienesas', 'Porotos', 'Tallarines', 'Papas fritas', 
                  'Estofado de carne', 'Carne de vacuno', 'Papas', 'Pastel de carne', 'Pollo', 'Zapatillo relleno', 
                  'Pastel de papas', 'Fideos espirales', 'Pollo apanado', 'Carne mongoliana', 'Pollo teriyaki', 
                  'Choclo a la crema', 'Strogonoff de pollo', 'Vegetariano', 'Salsa alfredo']

# Ruta base donde se encuentran las carpetas 'train', 'val', 'test'
path_base = 'data/juntas'
contar_etiquetas_por_clase(path_base, nombres_clases)
```
