## Total

```python
import torchvision.transforms as transforms
from torchvision.datasets import CocoDetection
from pycocotools.coco import COCO

ruta_imagenes_total = 'data/merge/integracion_datos/' 
ruta_annotations_total = 'data/merge/integracion_datos/_annotations.coco.json'

# Cargar el archivo annotations COCO
coco_total = COCO(ruta_annotations_total)

# Transformaciones de datos que deseas aplicar a las imágenes
transformaciones_total = transforms.Compose([
    transforms.ToTensor(),      # Convertir la imagen PIL a un tensor de PyTorch
    # Otras transformaciones que desees aplicar, como el reajuste del tamaño, normalización, etc.
])

# Cargar el conjunto de datos COCO usando CocoDetection
conjunto_datos_total = CocoDetection(root=ruta_imagenes_total, annFile=ruta_annotations_total, transform=transformaciones_total)
```

## Visualizacion de datos total

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np

#for i in range(len(conjunto_datos_total)):
for i in range(5):
    imagen, anotaciones = conjunto_datos_total[i]
    imagen_numpy = imagen.permute(1, 2, 0).numpy()
    plt.imshow(imagen_numpy)
    for anotacion in anotaciones:
        bbox = anotacion['bbox']
        segmentacion = anotacion['segmentation']
        categoria_id = anotacion['category_id']
        categoria_nombre = coco_total.loadCats(categoria_id)[0]['name']
        
        # Mostrar el bounding box
        x, y, width, height = bbox
        # rect = patches.Rectangle((x, y), width, height, linewidth=2, edgecolor='r', facecolor='none')
        # plt.gca().add_patch(rect)
        
        # Mostrar el nombre de la categoría cerca del rectángulo
        plt.text(x, y, categoria_nombre, fontsize=8, color='r')

        # Mostrar la segmentacion o mascara
        for segmento in segmentacion:
            puntos = np.array(segmento).reshape(-1, 2)
            poligono = patches.Polygon(puntos, closed=True, edgecolor='r', facecolor='none')
            plt.gca().add_patch(poligono)

    plt.show()
    print('Esto es el tensor de imagen: ', imagen)
    print('Esto es el tensor de anotaciones: ', anotaciones)
```
